```python
import pandas as pd
from IPython import display
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('hate_crime-csv-2-csv.csv')
data
```

```python
print ("Number of rows in the dataset  : " ,data.shape[0])
print ("Number of Columns in the dataset : " ,data.shape[1])
print ("Number of Features : \n" ,data.columns.tolist())
print ("Unique values :  \n",data.nunique())
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 30,
        }
l=data.columns.tolist()
for i in range(1,len(l)):
    
    
    if len(data.value_counts(l[i]))>50:
        
        df = data.value_counts(l[i])[:50]
        plt.figure(figsize=(30,10))
        sns.barplot(x=df.index, y=df)
        plt.xticks(rotation=90)
        plt.title(l[i],fontdict=font)
    
    else:
        
        df = data.value_counts(l[i])
        plt.figure(figsize=(30,10))
        sns.barplot(x=df.index, y=df)
        plt.xticks(rotation=90)
        plt.title(l[i],fontdict=font)
```
